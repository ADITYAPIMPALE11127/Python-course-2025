### üîπ Scatter Plot ‚Äî Key Points

### Purpose:

Shows relationship between two numeric variables.

Each point = one observation.

### Parameters:

x, y ‚Üí numeric columns.

hue ‚Üí color points by category.

size ‚Üí marker size by numeric variable.

style ‚Üí marker shape by category.

alpha ‚Üí transparency (useful for overlapping points).

### Variations:

sns.scatterplot ‚Üí plain scatter, customizable.

sns.regplot ‚Üí scatter + regression line (trend).

sns.lmplot ‚Üí scatter + regression, supports grouping (hue, col).

Extra Options:

fit_reg=False ‚Üí removes regression line in regplot/lmplot.

order=n ‚Üí polynomial regression (e.g., quadratic).

ci= ‚Üí controls confidence interval shading.

### üîπ Use Cases

Correlation check: e.g., budget vs box office gross.

Trend spotting: upward/downward relation between variables.

Cluster detection: see natural groupings.

Outlier detection: find unusual points away from the trend.

Multi-variable analysis: use hue, size, style for more dimensions.

‚ú® Quick memory hook:
Scatter plot = ‚ÄúDots for Dependence‚Äù ‚Üí dots show how one variable depends on another.

```python
import seaborn as sns
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = sns.load_dataset('iris')
```

```python
df.head()
```

### üìñ Theory

Scatter plots show relationship between two numeric variables.

Each point = one observation.

Useful to spot trends, clusters, correlations, and outliers.

hue ‚Üí add a categorical variable (points get different colors).

size ‚Üí vary marker size by a numeric column.

```python
sns.scatterplot(x ='sepal_length',y = 'petal_length',hue='species',data=df)
```

## Add size (marker size by numeric column)

```python
sns.scatterplot(x ='sepal_length',y = 'petal_length',size='petal_width',data=df)
```

## 3. Change style (marker shapes by category)

```python
sns.scatterplot(x='sepal_length',y='sepal_width',data=df,style='species',hue='species',palette='spring')
```

## Combine hue + size + style

üëâ Multi-dimensional view (color + size + shape).

```python
sns.scatterplot(x='petal_length',y='petal_width',hue='species',style='species',size='sepal_length',data=df)
```

## 5. Transparency for overlap

üëâ Avoid clutter when many points overlap.

```python
sns.scatterplot(x='sepal_width',y='sepal_length',alpha=0.5,data=df,hue='species',palette='winter_r')
```

## Regression Scatter Plots
1. sns.regplot

Plots scatter + regression line (by default, linear).

Best when you just want a single relationship.

```python
sns.regplot(x='sepal_length',y='sepal_length',data=df)
```

```python
sns.regplot(x="sepal_width", y="petal_width", data=df,
            scatter_kws={"s":40, "alpha":0.5})
```

```python
sns.lmplot(x="sepal_length", y="petal_length", hue="species", data=df)
```
