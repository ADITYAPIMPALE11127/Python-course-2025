### What is countplot?

seaborn.countplot is used to show the count of categories in a categorical variable.

Works like a histogram but for categorical data (instead of numeric bins).

Each bar = number of observations in that category.

```python
import numpy as np
import seaborn as sns
import pandas as pd
import matplotlib.pyplot as plt


%matplotlib inline
```

```python
df = pd.read_csv('imdb_movie_dataset.csv')
```

```python
df.head()
```

```python
sns.countplot(x='Year',hue='Rating',data=df,palette='spring')
plt.show()
```

```python
sns.countplot(data=df,x ='Rating')
plt.xticks(rotation=45)
plt.show()
```

```python
sns.countplot(x="Genre", data=df,
              order=df['Genre'].value_counts().index)
plt.xticks(rotation=45)
plt.show()
```

```python
ax = sns.countplot(x="Genre", data=df)
plt.xticks(rotation=45)

# Add labels on top of bars
for p in ax.patches:
    ax.annotate(p.get_height(), (p.get_x()+0.3, p.get_height()+5))
plt.show()
```

```python
sns.countplot(x="Genre", hue="Rating", data=df, dodge=False)
plt.xticks(rotation=45)
plt.show()
```
