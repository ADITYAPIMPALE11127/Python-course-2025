# Apply Method

```python
n1 = 410
```

```python
'*'*12+str(n1)[-4:]
```

```python
def mask_cc(n):
  return '*'*12+str(n)[-4:]
```

```python
mask_cc(4676137647685994)
```

```python
df['masked_cc'] = df['CC Number'].apply(mask_cc)
df.head()
```

```python
def cat_customer(amt):
  if amt < 10:
    return '$'
  elif amt >= 10 and amt <= 17:
    return '$$'
  else:
    return '$$$'
```

```python
df['cust_category'] = df['total_bill'].apply(cat_customer)
df.head()
```
